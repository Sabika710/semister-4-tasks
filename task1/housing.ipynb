```python
def fillnaobj(cols):
    for i in cols:
        df[i]=df[i].fillna(df[i].mode()[0])

columns = ['LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'ExterQual', 'ExterCond', 'Foundation','BsmtQual', 'BsmtCond', 'BsmtExposure','SaleType', 'SaleCondition', 'BsmtFinType1', 'HeatingQC', 'KitchenQual', 'RoofStyle', 'RoofMatl', 'Exterior1st','Exterior2nd']
fillnaobj(columns)
```

```python
def objtoint(cols):
    for i in cols:
        encoder= LabelEncoder()
        df[i] = encoder.fit_transform(df[i])
columns =['LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'ExterQual', 'ExterCond', 'Foundation','BsmtQual', 'BsmtCond', 'BsmtExposure','SaleType', 'SaleCondition', 'BsmtFinType1', 'HeatingQC', 'KitchenQual', 'RoofStyle', 'RoofMatl', 'Exterior1st','Exterior2nd']
objtoint(columns)
```

```python
def fillNaFloat(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mean())

columns=['MasVnrArea', 'BsmtFullBath', 'BsmtHalfBath', 'GarageCars', 'GarageArea'
        ]
fillNaFloat(columns)
```

```python
def convertfloat(cols):
    for i in cols:
        df[i]=df[i].astype("int64")

colunms = ['MasVnrArea', 'BsmtFullBath', 'BsmtHalfBath', 'GarageCars', 'GarageArea']
convertfloat(colunms)
```
