```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
```

```python
test = pd.read_csv("test.csv")
train = pd.read_csv("train.csv")
train
```

```python
train.drop(["Name","Cabin","PassengerId","RoomService","FoodCourt","ShoppingMall"], axis=1 , inplace = True)
test.drop(["Name","Cabin","PassengerId","RoomService","FoodCourt","ShoppingMall"], axis=1 , inplace = True)
```

```python
train.info()
```

```python
def fillobj(cols):
  for i in cols:
    train[i] = train[i].fillna(train[i].mode()[0])
col=["HomePlanet","CryoSleep","Destination","VIP"]
fillobj(col)
```

```python
def fillfloat(cols):
    for i in cols:
        train[i]=train[i].fillna(train[i].mean())
cols=["Age","Spa","VRDeck"]
fillfloat(cols)
```

```python
train.isnull().sum()
```

```python
train.info()
```

```python
def objtoint(col):
    encoder = LabelEncoder()
    for i in col:
        train[i]= encoder.fit_transform(train[i])
col=["HomePlanet","CryoSleep","Destination","VIP","Transported"]
objtoint(col)
```

```python
def floattoint(col):
    for i in col:
        train[i]=train[i].astype("int64")
cols=["Age","Spa","VRDeck"]       
floattoint(cols)
```

```python
train.info()
```

```python
def fillobj(cols):
  for i in cols:
    test[i] = test[i].fillna(test[i].mode()[0])
col=["HomePlanet","CryoSleep","Destination","VIP"]
fillobj(col)
```

```python
def fillfloat(cols):
    for i in cols:
        test[i]=train[i].fillna(test[i].mean())
cols=["Age","Spa","VRDeck"]
fillfloat(cols)
```

```python
def objtoint(col):
    encoder = LabelEncoder()
    for i in col:
        test[i]= encoder.fit_transform(test[i])
col=["HomePlanet","CryoSleep","Destination","VIP"]
objtoint(col)
```

```python
def floattoint(col):
    for i in col:
        test[i]=test[i].astype("int64")
cols=["Age","Spa","VRDeck"]       
floattoint(cols)
```

```python
train_x = train.drop("Transported",axis=1)
train_y = train["Transported"]
test_x = test
```

```python
model1 = RandomForestClassifier()
model1.fit(train_x, train_y)
```

```python
testdata = pd.read_csv('test.csv')

```

```python
predicted = model1.predict(test_x)
predicted
pridicted = np.where(predicted == 1, True,False)
```

```python
pridicted
```

```python
submission = pd.DataFrame({'PassengerId':testdata['PassengerId'],'Transported':pridicted})
submission.to_csv('submission.csv', index=False)
```
